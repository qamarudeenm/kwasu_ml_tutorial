# Topic: Image Data Preprocessing & Transformation

## Introduction: The "Garbage In, Garbage Out" Principle
In Machine Learning, a model is only as good as the data you feed it. A Convolutional Neural Network (CNN) is effectively a complex mathematical formula. If you feed it inconsistent data—images of different sizes, different lighting conditions, or massive number ranges—the formula cannot solve the problem.

**The Goal:** We preprocess data to make it **Uniform**, **Numerical**, and **Mathematically Stable**.

## Module 1: Geometric Transformations (Uniformity)

**The Problem: Dimension Mismatch**
Real-world images come in all shapes ($1920 \times 1080$, $500 \times 500$, etc.). However, the **Input Layer** of a CNN is rigid. It expects a fixed vector size (e.g., exactly $224 \times 224$ pixels). We must reshape all images to match this requirement.

**The Algorithm: Interpolation**
When we shrink or stretch an image, we change the grid of pixels.
*   **Shrinking:** We lose information and must decide which pixels to keep.
*   **Stretching:** We create space and must guess what colors go in the new gaps.
*   **Bilinear Interpolation:** The computer looks at a $2 \times 2$ block of original pixels and calculates a weighted average to generate the new pixel. This smoothens the image and prevents jagged edges.

## Module 2: Numerical Transformations (Stability)

**The Problem: The Scale of Numbers**
A standard digital image is stored as integers from **0 (Black)** to **255 (White)**.
However, Neural Networks use weights that are very small (e.g., $0.001$, $0.05$).

**The Consequence: Exploding Gradients**
If you multiply a large input ($255$) by a weight ($0.05$) inside a deep network with many layers, the numbers can spiral out of control. This makes the model unstable and unable to learn.

**The Solution: Normalization (Min-Max Scaling)**
We force every pixel to be between **0.0 and 1.0**.
$$ X_{new} = \frac{X_{current}}{255.0} $$
This keeps the math stable and helps the Optimizer (e.g., Gradient Descent) converge faster.

## Module 3: Feature Engineering (Data Augmentation)

**The Theory: Invariance**
In traditional ML (like Excel data), Feature Engineering means creating new columns. In Image ML, it means **Augmentation**.
A good AI model should understand that an object is the same regardless of its orientation or position.
*   **Rotational Invariance:** A cat upside down is still a cat.
*   **Translation Invariance:** A cat on the left or right is still a cat.

**The Solution: Artificial Data Generation**
By randomly rotating, flipping, zooming, or shifting our training images, we create "Free Data." This forces the model to learn the *features* (ears, whiskers) rather than memorizing the specific *pixels* of a single photo.

## Module 4: Label Engineering (Target Transformation)

**The Problem: False Ordinality**
We must also process the answers (Labels). Imagine we have classes: `0=Cat`, `1=Dog`, `2=Bird`.
If we feed these numbers directly to the model, the math assumes:
*   $Bird (2) > Cat (0)$
*   $Average(Cat, Bird) = Dog$

**This is false logic.** These are categories, not rankings.

**The Solution: One-Hot Encoding**
We convert the single integer into a binary vector.
*   Cat (0) becomes: `[1, 0, 0]`
*   Dog (1) becomes: `[0, 1, 0]`
*   Bird (2) becomes: `[0, 0, 1]`

# Part 2: The Practice (Lab Phase)


```python
# STEP 0: SETUP
# Import necessary libraries for the lab
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.utils import to_categorical

# Configuration for visualization
%matplotlib inline
print(f"Setup complete. TensorFlow Version: {tf.__version__}")
```

```python
# STEP 1: CREATE RAW DATA
# Since we don't have a file, we will create a "Dummy" 100x100 pixel image.
# We add squares of Red, Green, and Blue to visualize changes easily.

raw_image = np.zeros((100, 100, 3), dtype='uint8')
raw_image[:50, :50, 0] = 255     # Top-Left Red
raw_image[50:, 50:, 1] = 255     # Bottom-Right Green
raw_image[40:60, 40:60, 2] = 255 # Center Blue

plt.figure(figsize=(3,3))
plt.imshow(raw_image)
plt.title(f"Original Raw Image\nShape: {raw_image.shape}")
plt.axis('off')
plt.show()
```

```python
# STEP 2: GEOMETRIC TRANSFORMATION (Resizing)
# We shrink the 100x100 image down to 50x50 using Bilinear Interpolation.

resized_image = tf.image.resize(raw_image, [50, 50], method='bilinear')

# Visualization comparison
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(raw_image)
plt.title(f"Original: {raw_image.shape}")
plt.axis('off')

plt.subplot(1, 2, 2)
# Cast to uint8 for display purposes
plt.imshow(resized_image.numpy().astype("uint8"))
plt.title(f"Resized: {resized_image.shape}")
plt.axis('off')
plt.show()
```

```python
# STEP 3: NUMERICAL TRANSFORMATION (Normalization)
# We convert values from [0, 255] (Integer) to [0, 1] (Float).

# 1. Check original
print(f"Original Max Value: {np.max(raw_image)}")

# 2. Convert to Float32
float_image = tf.cast(raw_image, tf.float32)

# 3. Normalize (Divide by 255)
normalized_image = float_image / 255.0

# 4. Check result
print(f"Normalized Max Value: {np.max(normalized_image)}")
print("The image data is now mathematically safe for the neural network.")
```

```python
# STEP 4: DATA AUGMENTATION (Feature Engineering)
# We apply random transformations to simulate "new" data.

# 1. Flip Left/Right
aug_flip = tf.image.flip_left_right(raw_image)

# 2. Adjust Saturation (Color intensity)
aug_sat = tf.image.adjust_saturation(raw_image, 3)

# 3. Adjust Brightness
aug_bright = tf.image.adjust_brightness(raw_image, delta=0.4)

# Visualization
plt.figure(figsize=(12, 4))

plt.subplot(1, 4, 1)
plt.imshow(raw_image)
plt.title("Original")
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(aug_flip)
plt.title("Augmented: Flip")
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(aug_sat)
plt.title("Augmented: Saturation")
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(aug_bright)
plt.title("Augmented: Brightness")
plt.axis('off')

plt.show()
```

```python
# STEP 5: LABEL ENGINEERING (One-Hot Encoding)
# We convert categorical integers into binary vectors.

# 1. Define raw labels (0=Cat, 1=Dog, 2=Bird)
raw_labels = [0, 1, 2, 0, 2]
print(f"Raw Labels: {raw_labels}")

# 2. Apply One-Hot Encoding
one_hot_labels = to_categorical(raw_labels, num_classes=3)

print("\nOne-Hot Encoded Labels:")
print(one_hot_labels)

# Notice how the values map to the position in the vector.
```

# Real Image
## "Now that we know the theory, let's pretend we are building a real Dog Classifier. We will take a random photo of a dog from the internet and prepare it for our neural network."


## Step 0: Setup & Image Loading
We use a utility to grab a reliable test image from the web.

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.utils import to_categorical
import requests # <--- New import

# 1. The Wikimedia URL that was blocking you
image_url = "https://upload.wikimedia.org/wikipedia/commons/thumb/4/47/American_Eskimo_Dog.jpg/600px-American_Eskimo_Dog.jpg"

# 2. Download manually using 'requests' with a fake User-Agent
# This tricks the server into thinking we are a Mozilla browser, not a script
headers = {'User-Agent': 'Mozilla/5.0'} 
response = requests.get(image_url, headers=headers)

# Save the file locally
local_path = "dog_image.jpg"
with open(local_path, "wb") as f:
    f.write(response.content)

# 3. Load the image from the local file we just saved
raw_image = tf.keras.utils.load_img(local_path)
raw_image = tf.keras.utils.img_to_array(raw_image).astype('uint8')

# Visualize
plt.figure(figsize=(5, 5))
plt.imshow(raw_image)
plt.title(f"Raw Input Image\nOriginal Size: {raw_image.shape}")
plt.axis('off')
plt.show()
```

## Step 1: Geometric Transformation (Resizing)
Real cameras take huge photos (e.g., 4000x3000). CNNs need small squares (e.g., 224x224).

```python
# THEORY APPLICATION:
# We need to squash this rectangular dog into a square box for the AI.

# 1. Define target size (Standard for many CNNs like ResNet/VGG)
target_size = (224, 224)

# 2. Apply Resizing with Bilinear Interpolation
# This calculates the average of pixels so the dog doesn't look "pixelated"
resized_image = tf.image.resize(raw_image, target_size, method='bilinear')

# Visualization
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(raw_image)
plt.title(f"Original\n{raw_image.shape}")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(resized_image.numpy().astype("uint8"))
plt.title(f"Preprocessed Input\n{resized_image.shape}")
plt.axis('off')

plt.show()
```

## Step 2: Numerical Transformation (Normalization)
Look at the pixel values change from "0-255" to "0-1".

```python
# THEORY APPLICATION:
# Neural networks hate large numbers. We must scale the pixel values.

# 1. Check a specific pixel (e.g., the dog's nose) before normalization
# We look at the center pixel
center_pixel = resized_image[112, 112]
print(f"Pixel value BEFORE normalization (RGB): {center_pixel.numpy()}")

# 2. Normalize
# We divide by 255.0 to get values between 0 and 1
normalized_image = resized_image / 255.0

# 3. Check the same pixel after
print(f"Pixel value AFTER normalization: {normalized_image[112, 112].numpy()}")
print(f"\nMax value in image: {np.max(normalized_image)}") # Should be 1.0 (or close to it)
```

## Step 3: Feature Engineering (Data Augmentation)
We generate "new" training samples from this single dog photo.
code


```python
# THEORY APPLICATION:
# If we rotate the dog, is it still a dog? Yes.
# We create variations to teach the AI invariance.

# 1. Create a "Batch" of images
# (Augmentation layers expect a batch, so we add a dimension: 1, 224, 224, 3)
batched_img = tf.expand_dims(resized_image, 0)

# 2. Define our Augmentation Pipeline
data_augmentation = tf.keras.Sequential([
  tf.keras.layers.RandomFlip("horizontal"),
  tf.keras.layers.RandomRotation(0.2), # Rotate up to 20%
  tf.keras.layers.RandomZoom(0.2),     # Zoom in/out up to 20%
])

# 3. Generate 3 different versions of the same dog
plt.figure(figsize=(12, 4))
for i in range(3):
    augmented_image = data_augmentation(batched_img)
    
    plt.subplot(1, 3, i + 1)
    plt.imshow(augmented_image[0] / 255.0) # Divide by 255 to display correctly
    plt.title(f"Augmentation {i+1}")
    plt.axis("off")

plt.suptitle("Synthesizing New Training Data")
plt.show()
```

## Step 4: Label Engineering (One-Hot Encoding)
Preparing the answer key.

```python
# THEORY APPLICATION:
# The computer cannot read the word "Dog". We must convert it to a vector.

# Let's say our classes are:
class_names = ['Cat', 'Dog', 'Bird', 'Horse']

# Our image is a Dog, which is index 1
label_index = 1 

# Apply One-Hot Encoding
one_hot = to_categorical(label_index, num_classes=len(class_names))

print(f"Class Name: {class_names[label_index]}")
print(f"Class Index: {label_index}")
print(f"One-Hot Vector: {one_hot}")

# Visualization of the vector
plt.figure(figsize=(6, 1))
plt.imshow([one_hot], cmap='gray')
plt.yticks([])
plt.xticks(range(4), class_names)
plt.title("Visual representation of the Label")
plt.show()
```
